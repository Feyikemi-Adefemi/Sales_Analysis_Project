```python
sales_by_category = df.groupby("Category")["Sales"].sum()

sales_by_category
	
# ============================================
# SALES BY CATEGORY
# Business Question:
# Which product category generated the highest sales?
# ============================================

sales_by_category = df.groupby("Category")["Sales"].sum()
print("Sales by Category")
print(sales_by_category)
print()


# ============================================
# SALES BY YEAR
# Business Question:
# How did sales change each year?
# ============================================

sales_by_year = df.groupby(df["Order Date"].dt.year)["Sales"].sum()
print("Sales by Year")
print(sales_by_year)
print()


# ============================================
# TOP 10 CUSTOMERS
# Business Question:
# Which customers generated the highest sales?
# ============================================

top_customers = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 Customers")
print(top_customers)
print()


# ============================================
# TOP 10 PRODUCTS
# Business Question:
# Which products generated the highest sales?
# ============================================

top_products = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 Products")
print(top_products)
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
# Sales by Region

region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))

plt.bar(region_sales.index, region_sales.values)

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales ($)")

plt.tight_layout()

plt.savefig("../Images/sales_by_region.png")

plt.show()
```

```python
# ==========================================
# SALES BY CATEGORY
# ==========================================

category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.tight_layout()
plt.savefig("../Images/sales_by_category.png")
plt.show()


# ==========================================
# SALES TREND BY YEAR
# ==========================================

sales_by_year = df.groupby(df["Order Date"].dt.year)["Sales"].sum()

plt.figure(figsize=(8,5))
plt.plot(sales_by_year.index,
         sales_by_year.values,
         marker="o",
         linewidth=2)

plt.title("Sales Trend (2015-2018)")
plt.xlabel("Year")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.tight_layout()
plt.savefig("../Images/sales_trend.png")
plt.show()


# ==========================================
# TOP 10 CUSTOMERS
# ==========================================

top_customers = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
plt.barh(top_customers.index,
         top_customers.values)

plt.title("Top 10 Customers")
plt.xlabel("Sales ($)")
plt.tight_layout()
plt.savefig("../Images/top_customers.png")
plt.show()


# ==========================================
# TOP 10 PRODUCTS
# ==========================================

top_products = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))
plt.barh(top_products.index,
         top_products.values)

plt.title("Top 10 Products")
plt.xlabel("Sales ($)")
plt.tight_layout()
plt.savefig("../Images/top_products.png")
plt.show()


print("========================================")
print(" All charts have been created!")
print(" Images saved successfully.")
print("========================================")
```
